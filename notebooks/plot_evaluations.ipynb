```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set an academic style (non-LaTeX for compatibility)
sns.set_context("paper", font_scale=1.2)
sns.set_style("whitegrid")
plt.rc('font', family='serif')
plt.rcParams.update({
    'font.size': 12,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'lines.linewidth': 1.5,
    'lines.markersize': 6,
    'legend.fontsize': 10,
    'figure.dpi': 300,
})

# Episodes (1 to 10)
episodes = list(range(1, 11))

double_dqn_rewards = [
    480.0,                # Episode 1
    335.2270259857178,     # Episode 2
    411.2874755859375,     # Episode 3
    312.624267578125,      # Episode 4
    414.9627914428711,     # Episode 5
    480.0,                # Episode 6
    299.15920639038086,    # Episode 7
    480.0,                # Episode 8
    480.0,                # Episode 9
    480.0 
]
double_dqn_avg = 417.33
double_dqn_std = 71.66

# Provided DuelingDoubleDQN evaluation data
dueling_rewards = [
    340.77623748779297,
    414.9627914428711,
    414.9627914428711,
    411.2874755859375,
    414.9627914428711,
    411.2874755859375,
    323.3337650299072,
    281.6095161437988,
    414.9627914428711,
    480.0
]
dueling_avg = 390.81
dueling_std = 54.87

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# DoubleDQN Evaluation Plot (Left)
ax1.plot(episodes, double_dqn_rewards, marker='o', linestyle='-', color='#0072B2', label='Episode Reward')
ax1.axhline(double_dqn_avg, color='black', linestyle='--', linewidth=1.5, label=f'Average = {double_dqn_avg:.2f}')
ax1.fill_between(episodes,
                 [double_dqn_avg - double_dqn_std] * len(episodes),
                 [double_dqn_avg + double_dqn_std] * len(episodes),
                 color='#0072B2', alpha=0.15, label=f'Std = {double_dqn_std:.2f}')
ax1.set_xlabel('Episode')
ax1.set_ylabel('Reward')
ax1.set_title('DoubleDQN Evaluation (0-$\\epsilon$)')
ax1.set_xticks(episodes)
ax1.legend(loc='best')
ax1.grid(True, linestyle='--', linewidth=0.5)

# DuelingDoubleDQN Evaluation Plot (Right) with a purplish color
purplish_color = '#8A2BE2'
ax2.plot(episodes, dueling_rewards, marker='o', linestyle='-', color=purplish_color, label='Episode Reward')
ax2.axhline(dueling_avg, color='black', linestyle='--', linewidth=1.5, label=f'Average = {dueling_avg:.2f}')
ax2.fill_between(episodes,
                 [dueling_avg - dueling_std] * len(episodes),
                 [dueling_avg + dueling_std] * len(episodes),
                 color=purplish_color, alpha=0.15, label=f'Std = {dueling_std:.2f}')
ax2.set_xlabel('Episode')
ax2.set_title('DuelingDoubleDQN Evaluation (0-$\\epsilon$)')
ax2.set_xticks(episodes)
ax2.legend(loc='best')
ax2.grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.savefig('evaluation_plots.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
